# Функции

Чтобы избежать повторного написания одной и той же логики, в языках программирования существуют функции.

Функции — это такие участки кода, которые изолированы от остальный программы и выполняются только тогда, когда вызываются. Вы уже встречались с функциями min(), len() и print(). Они все обладают общим свойством: они могут принимать параметры (ноль, один или несколько), и они могут возвращать значение (хотя могут и не возвращать). Например, функция min() принимает один параметр (например, список) и возвращает значение - минимальное число. Функция print() принимает переменное число параметров и ничего не возвращает.

Покажем, как написать функцию factorial(), которая принимает один параметр — число, и возвращает значение — факториал этого числа.


Напомним, что в математике факториал числа n определяется как n! = 1 ⋅ 2 ⋅ ... ⋅ n. Например, 5! = 1 ⋅ 2 ⋅ 3 ⋅ 4 ⋅ 5 = 120. Ясно, что факториал можно легко посчитать, воспользовавшись циклом for. Представим, что нам нужно в нашей программе вычислять факториал разных чисел несколько раз (или в разных местах кода). 

```python
# вычислим 3!
res = 1
for i in range(1, 4):
    res *= i
print(res)

# вычислим 5!
res = 1
for i in range(1, 6):
    res *= i
print(res)
```

Вместо копирования фрагмента кода и исправления конкретных значений переменных (например, параметров range), можно описать этот код в "общем случае", используя формальный параметр n.

```python
def factorial(n):
    res = 1
    for i in range(1, n + 1):
        res *= i
    return res
```

а затем вызывать этот код, передавая конкретное значение параметра

```python
print(factorial(3))
print(factorial(5))
```

Код функции должен размещаться выше того того места, где мы захотим воспользоваться функцией factorial().  factorial — идентификатор, то есть имя нашей функции. После идентификатора в круглых скобках идет список параметров, которые получает наша функция. Список состоит из перечисленных через запятую идентификаторов параметров. В нашем случае список состоит из одной величины n. В конце строки ставится двоеточие.

Далее идет тело функции, оформленное в виде блока, то есть с отступом. Внутри функции вычисляется значение факториала числа n и оно сохраняется в переменной res. Функция завершается инструкцией return res, которая завершает работу функции и возвращает значение переменной res.


Инструкция return завершает работу функции и возвращает указанное значение в место вызова. Если функция не возвращает значения, то инструкция return используется без возвращаемого значения. В функциях, которым не нужно возвращать значения, инструкция return может отсутствовать.



```python
def my_function(name):
  print("Hello, ", name, "!")

  

my_function("Emil")
my_function("Tobias")
my_function("Linus")
```

# Передача нескольких параметров

Параметров может быть несколько, при вызове функции в этом случае важен порядок аргументов. Например, напишем функцию max(), которая принимает два числа и возвращает максимальное из них (на самом деле, такая функция уже встроена в Python, но в этой задаче мы не будем ее использовать).

```python
def max(a, b):
    if a > b:
        return a
    else:
        return b

print(max(3, 5)) 

x = 100
y = 200
print(max(x, y))

print(max(int(input()), int(input())))
```

Теперь можно написать функцию max3(), которая принимает три числа и возвращает максимальное их них.

```python
def max(a, b):
    if a > b:
        return a
    else:
        return b

def max3(a, b, c):
    return max(max(a, b), c)

print(max3(3, 5, 4))
```

переменные, которые используются в описании функции называются **параметрами** функции (формальными параметрами), а переменные, использующиеся при вызове функции называются **аргументами** (фактическими параметрами)

```python
#описание функции
#child1 child2 - параметры функции

def my_child(child1, child2):
  print("Мы любим наших детей " + child1 +  " и " + child2)

#вызов функции
# "Emil", "Tobias" - аргументы

my_child("Emil", "Tobias")
```

## Произвольное количество рагументов, *args

Если вы не знаете, сколько аргументов будет передано в вашу функцию, добавьте * перед именем параметра в определении функции.

Таким образом, функция получит кортеж аргументов и сможет получить доступ к элементам этого кортежа:

```python
def all_kids(*kids):
  result = ''
  for child in kids:
    result = result + " " + child +", "
  print("Все наши дети: " + result + "они умненькие")

all_kids("Emil", "Tobias", "Linus", "Ната")
```

>в документации Python произвольное количество аргументов часто сокращаются до *args

## Передача аргументов с ключевыми словами (именами параметров)

Передавать аргументы при вызове функции можно в форме ключ = значение, где ключ - это имя параметра. в этом случае порядок аргументов не имеет значение

```python
def my_function(child1, child2, child3):
  print("The youngest child is " + child3)

my_function(child3 = "Emil", child1 = "Tobias", child2 = "Linus")
```

## Передача произвольного количества аргументов с ключевыми кловами, **kwargs

Если вы не знаете, сколько аргументов ключевого слова будет передано в вашу функцию, добавьте две звездочки: ** перед именем параметра в определении функции.
Таким образом, функция получит словарь аргументов и сможет соответственно получить доступ к элементам:

```python
def my_function(**kid):
  print("His last name is " + kid["lname"])

my_function(fname = "Tobias", lname = "Refsnes")
```

> в документации Python произвольное количество аргументов, передаваемых с ключевыми словами часто сокращаются до **kwargs .

Квадратный корень можно вычислить с помощью возведения в степень 1/2

```python
def distance(x1, y1, x2, y2):
    return ((x2 - x1) ** 2 + (y2 - y1) ** 2) ** 0.5

if __name__ == "__main__":
    x1 = float(input("Введите координату x1: "))
    y1 = float(input("Введите координату y1: "))
    x2 = float(input("Введите координату x2: "))
    y2 = float(input("Введите координату y2: "))

    dist = distance(x1, y1, x2, y2)

    print(f"Расстояние между точками ({x1}, {y1}) и ({x2}, {y2}) равно {dist:.2f}")
```

### 2. Доработайте функцию distance так, чтобы ее можно было вызвать только с двумя аргументами distance(x,y).  В таком случае эта функция должна возвращать расстояние от точки с координатами (x,y) до начала координат (0,0). Попросите пользователя ввести координаты двух точек и определите какая из точек находится ближе к началу координат 

```python
def distance(x1, y1, x2=None, y2=None):
    if x2 is None and y2 is None:
        return (x1 ** 2 + y1 ** 2) ** 0.5
    else:
        return ((x2 - x1) ** 2 + (y2 - y1) ** 2) ** 0.5

if __name__ == "__main__":
    x1 = float(input("Введите координату x1: "))
    y1 = float(input("Введите координату y1: "))

    x2 = float(input("Введите координату x2: "))
    y2 = float(input("Введите координату y2: "))

    dist1 = distance(x1, y1)
    dist2 = distance(x2, y2)

    print(f"Расстояние от точки ({x1}, {y1}) до начала координат равно {dist1:.2f}")
    print(f"Расстояние от точки ({x2}, {y2}) до начала координат равно {dist2:.2f}")

    if dist1 < dist2:
        print(f"Точка ({x1}, {y1}) ближе к началу координат.")
    elif dist1 > dist2:
        print(f"Точка ({x2}, {y2}) ближе к началу координат.")
    else:
        print("Обе точки находятся на одинаковом расстоянии от начала координат.")
```

 3. Напишите функцию my_norm_str(), которая принимает слово из букв разного регистра (заглавных и строчных) и возвращает то же слово, в котором все буквы, кроме первой - строчные, а первая буква - заглавная.
Например, print(my_norm_str('wOrD')) должно печатать слово Word.

На вход подаётся строка, состоящая из слов, разделённых одним пробелом. Слова состоят из букв разного регистра. Напечатайте исходную строку, сделав так, чтобы каждое слово начиналось с большой буквы. При этом используйте вашу функцию my_norm_str().
>в Python для строк определен метод  title()  - преобразует первую букву каждого слова к верхнему регистру. Не используйте его в этой задаче.

```python
def my_norm_str(word):
    if len(word) == 0: 
        return word
    return word[0].upper() + word[1:].lower()

if __name__ == "__main__":
    input_string = input("Введите строку: ")
    
    normalized_words = [my_norm_str(word) for word in input_string.split()]
    
    normalized_string = ' '.join(normalized_words)

    print(normalized_string)
```

### 7.
Напишите функцию, которая по номеру месяца возвращает количество дней в этом месяце. Считается, что год невисокосный. Продемонстрируйте работу этой функции в скрипте, в котором пользователей вводит номер месяца с клавиатуры. При вводе неверного номера месяца должно быть выведено сообщение об ошибке.

```python
def days_in_month(month):
    if month < 1 or month > 12:
        return "Ошибка: неверный номер месяца"
    
    days = [31, 28, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31]
    return days[month - 1]

if __name__ == "__main__":
    month = int(input("Введите номер месяца (1-12): "))
    result = days_in_month(month)
    print(result)
```

### 8.
Напишите функцию, которая определяет, является ли год високосным. Решите предыдущую задачу, с учетом того, что в високосном году в феврале 29 дней (если вводится номер месяца 2, то дополнительно просят ввести номер года). Для этого измените функцию, возвращающую количество дней в месяце: у нее должен быть необязательный параметр is_leap (по умолчанию False)

```python
def days_in_month(month, is_leap=False):
    if month < 1 or month > 12:
        return "Ошибка: неверный номер месяца"
    
    days = [31, 28, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31]
    
    if month == 2 and is_leap:
        return 29
    
    return days[month - 1]

def is_leap_year(year):
    return (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0)

if __name__ == "__main__":
    month = int(input("Введите номер месяца (1-12): "))
    
    if month == 2:
        year = int(input("Введите год: "))
        leap = is_leap_year(year)
        days = days_in_month(month, is_leap=leap)
    else:
        days = days_in_month(month)
    
    print(days)
```
